# Benchmark

```python
import os
import subprocess
import shutil

GDRIVE_DATASETS_PATH = "/content/drive/MyDrive/MASt3R-SLAM"
GDRIVE_RESULTS_PATH = "/content/drive/MyDrive/MASt3R-SLAM/results"
LOCAL_DATA_PATH = "/content/datasets"
MAST3R_SLAM_PATH = "/content/MASt3R-SLAM"
NUM_RUNS = 5

datasets_to_process = {
    #"tartanair.zip": {"tartanair": ["visual_endofworld", "visual_moon", "visual_westerndesert"]},
    #"rover.zip": {"rover": ["garden_small_2024-01-13-winter", "garden_small_2024-05-29_1-day", "garden_small_2024-05-29_4-night"]},
    #"kitti.zip": {"kitti": ["00", "01", "03"]},
    "euroc.zip": {"euroc": ["MH_04_difficult", "MH_05_difficult", "V1_03_difficult", "V2_03_difficult"]},
    "tum.zip": {"tum": ["rgbd_dataset_freiburg1_desk", "rgbd_dataset_freiburg1_plant", "rgbd_dataset_freiburg1_rpy", "rgbd_dataset_freiburg1_xyz"]},
    # "uzhfpv.zip": {"uzhfpv": ["..."]}
}


def get_paths_for_sequence(dataset_name, seq_name, local_dataset_root):
    dataset_path = None
    calib_path = None

    if dataset_name == "kitti":
        dataset_path = os.path.join(local_dataset_root, "data_odometry_color", seq_name, "image_2")
        if seq_name in ["00", "01"]:
            calib_path = os.path.join(local_dataset_root, "mast3rslam-intrinsics-00-02copy.yaml")
        else:
            calib_path = os.path.join(local_dataset_root, "mast3rslam-intrinsics-03.yaml")

    elif dataset_name == "rover":
        dataset_path = os.path.join(local_dataset_root, seq_name, "pi_cam", "rgb")
        calib_path = os.path.join(local_dataset_root, "calibration", "mast3rslam-intrinsics.yaml")

    elif dataset_name == "tartanair":
        dataset_path = os.path.join(local_dataset_root, seq_name, "image_lcam_front")
        calib_path = os.path.join(local_dataset_root, "mast3rslam-intrinsics.yaml")

    elif dataset_name == "euroc":
        dataset_path = os.path.join(local_dataset_root, seq_name)
        calib_path = os.path.join(local_dataset_root, "mast3rslam-intrinsics.yaml")

    elif dataset_name == "tum":
        dataset_path = os.path.join(local_dataset_root, seq_name)
        calib_path = os.path.join(local_dataset_root, "mast3rslam-intrinsics.yaml")

    else:
        raise ValueError(f"Dataset desconhecido: {dataset_name}")

    return dataset_path, calib_path


def run_command(command):
    process = subprocess.Popen(command, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True, cwd=MAST3R_SLAM_PATH)
    for line in process.stdout:
        print(line, end='')
    process.wait()
    if process.returncode != 0:
        print(f"ERRO: O comando retornou o código de saída {process.returncode}")

os.makedirs(LOCAL_DATA_PATH, exist_ok=True)
os.makedirs(GDRIVE_RESULTS_PATH, exist_ok=True)

for zip_file, info in datasets_to_process.items():
    dataset_name = list(info.keys())[0]
    sequences = list(info.values())[0]

    print(f"PROCESSANDO DATASET: {dataset_name.upper()}")

    zip_path = os.path.join(GDRIVE_DATASETS_PATH, zip_file)
    local_dataset_root = os.path.join(LOCAL_DATA_PATH, dataset_name)

    if os.path.exists(local_dataset_root):
        shutil.rmtree(local_dataset_root)

    if os.path.exists(zip_path):
        shutil.unpack_archive(zip_path, LOCAL_DATA_PATH)
    else:
        print(f"Arquivo {zip_file} não encontrado. Pulando.")
        continue

    for seq in sequences:
        results_dir_seq = os.path.join(GDRIVE_RESULTS_PATH, dataset_name, seq)
        os.makedirs(results_dir_seq, exist_ok=True)

        for i in range(1, NUM_RUNS + 1):
            print(f"Rodada {i}/{NUM_RUNS}...")
            logs_folder = os.path.join(MAST3R_SLAM_PATH, "logs")
            if os.path.exists(logs_folder):
                shutil.rmtree(logs_folder)

            dataset_path, calib_path = get_paths_for_sequence(dataset_name, seq, local_dataset_root)

            if not os.path.exists(calib_path):
                print(f"ERRO: Arquivo de calibração nao encontrado em: {calib_path} \n Pulando...")
                continue

            command = ["python", "main.py", "--no-viz", "--dataset", dataset_path, "--config", "config/base.yaml", "--calib", calib_path]
            run_command(command)

            source_traj_file = os.path.join(MAST3R_SLAM_PATH, "logs", f"{seq}.txt")
            source_ply_file = os.path.join(MAST3R_SLAM_PATH, "logs", f"{seq}.ply")

            if os.path.exists(source_traj_file):
                dest_traj_file = os.path.join(results_dir_seq, f"trajectory_run_{i}.txt")
                shutil.copy(source_traj_file, dest_traj_file)
            else:
                print(f"Arquivo de trajetória .txt não encontrado para a rodada {i}.")

            if os.path.exists(source_ply_file):
                dest_ply_file = os.path.join(results_dir_seq, f"reconstruction_run_{i}.ply")
                shutil.copy(source_ply_file, dest_ply_file)
            else:
                print(f"Arquivo de reconstrução .ply não encontrado para a rodada {i}.")

    shutil.rmtree(local_dataset_root)
```
